# Section 4.7

```python
weight = 0.5
input = 0.5
goal_prediction = 0.8
step_amount = 0.001

for iteration in range(1001):
    prediction = input * weight
    error = (goal_prediction - prediction) ** 2
    print('Error: {} Prediction: {}'.format(error, prediction))
    
    up_prediction = input * (weight + step_amount)
    up_error = (goal_prediction - up_prediction) ** 2
    
    down_prediction = input * (weight - step_amount)
    down_error = (goal_prediction - down_prediction) ** 2
    
    if down_error < up_error:
        weight -= step_amount
    
    if up_error < down_error:
        weight += step_amount
    
```

# Section 4.9

```python
weight = 0.5
input = 0.5
goal_pred = 0.8

for iteration in range(20):
    pred = input * weight
    error = (goal_prediction - pred) ** 2
    direction_and_amount = (pred - goal_pred) * input
    weight -= direction_and_amount
    print('Error: {} Prediction: {}'.format(error, pred))
```

# Section 4.10

```python
weight = 0.1
alpha = 0.1
def neural_network(input, weight):
    prediction = input * weight
    return prediction

number_of_toes = [8.5]
win_or_lose_binary = [1]

input = number_of_toes[0]
goal_pred = win_or_lose_binary[0]

pred = neural_network(input, weight)
error = (pred - goal_pred) ** 2

delta = pred - goal_pred
weight_delta = delta * input
weight -= weight_delta * alpha
```

# Section 4.11

```python
weight, goal_pred, input = (0.0, 0.8, 0.5)
for iteration in range(4):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * input
    weight -= weight_delta * alpha
    print('Error: {}, Prediction: {}'.format(error, pred))
```

# Section 4.12

```python
weight, goal_pred, input = (0.0, 0.8, 1.1)
for iteration in range(4):
    print('-----\nWeight: {}'.format(weight))
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * input
    weight -= weight_delta
    print('Error: {}, Prediction: {}'.format(error, pred))
    print('Delta: {}, Weight Delta: {}'.format(delta, weight_delta))
```

# Section 4.21

```python
weight = 0.5
goal_pred = 0.8
input = 2.0

weights = []
errors = []
for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = input * delta
    weight = weight - weight_delta
    print("Error:" + str(error) + " Prediction:" + str(pred))
    errors.insert(0, error)
    weights.insert(0, weight)
```

## Plot Error

```python
fig, axes = plt.subplots()
line = axes.plot(weights, errors)
axes.set_xlabel("weight value")
axes.set_ylabel("error value")
fig.show()
```

# Section 4.25

```python
weight = 0.5
goal_pred = 0.8
input = 2
alpha = 0.1
weights = []
errors = []
for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    derivative = input * (pred - goal_pred)
    weight = weight - (alpha * derivative)
    print("Error:" + str(error) + " Prediction:" + str(pred))
    errors.insert(0, error)
    weights.insert(0, weight)
print(errors)
```

```python
fig, axes = plt.subplots()
line = axes.plot(weights, errors)
axes.set_xlabel("weight value")
axes.set_ylabel("error value")
fig.show()
```
